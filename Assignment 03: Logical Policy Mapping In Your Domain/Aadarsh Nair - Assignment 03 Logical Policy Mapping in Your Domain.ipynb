## Section 3 — Classifying Logical Constraint Types

Policy 1

- Rule: If discounts are offered, then online ads must be run.
- Type: Activation
- Explanation: Choosing discounts requires online advertising.

Policy 2

- Rule: If electric vehicles are stocked, then financing must be offered.
- Type: Activation
- Explanation: EV sales depend on financing availability.

Policy 3

- Rule: If open weekends, then hire extra staff.
- Type: Activation
- Explanation: Weekend operations require additional staffing.

Policy 4

- Rule: If discounts are offered, financing cannot be offered.
- Type: Suppression
- Explanation: Discounts prevent financing from being selected.

Policy 5

- Rule: At least one of SUVs or EVs must be stocked.
- Type: Either-or
- Explanation: The dealership must choose at least one inventory type.
- Clarification: This is not “exactly one,” both could be allowed unless restricted elsewhere.

Policy 6

- Rule: If SUVs are stocked, EVs cannot be stocked.
- Type: Mutually exclusive
- Explanation: Only one inventory type can be selected due to space limits.

Policy 7

- Rule: If online ads are run, discounts must be offered.
- Type: Activation
- Explanation: Advertising campaigns are tied to promotional discounts.

## Section 5 — Feasibility, Conflicts, and "No Feasible Solution"


A major conflict arises between Policy 2 (EVs require financing) and Policy 4 (discounts prevent financing) combined with Policy 7 (ads require discounts). If the dealership wants to stock EVs and run ads, it creates a chain where financing is required but also prohibited.

This could make the model infeasible because no combination of decisions satisfies all policies simultaneously. For example, stocking EVs requires financing, but running ads requires discounts, which then prohibit financing—creating a contradiction.

In practice, “no feasible solution” would look like a manager trying to finalize a sales plan but finding that every option violates at least one policy. For instance, they cannot both promote vehicles and stock EVs without breaking financing or discount rules.

The first policy to relax would likely be Policy 4 (discount vs. financing restriction), as real dealerships often offer both under certain conditions. This policy is overly rigid compared to real-world flexibility.

To explain this to a non-technical stakeholder, I would say:
“The current rules conflict with each other. Some policies require financing while others prohibit it under the same conditions. Because of this, there is no possible plan that satisfies all rules. We need to adjust at least one policy to make a workable plan.”
